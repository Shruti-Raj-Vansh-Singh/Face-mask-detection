To train the model we have to run train.py, which takes the following arguments:

   1. img: input image size
   2. batch: batch size
   3. epochs: number of epochs
   4. data: path to the data.yml file
   5. cfg: model to choose among the preexisting in 📁models
   6. weights: initial weights path, defaults to yolov5s.pt
   7. name: renames output folder
   8. device: Wheter to train on cpu or gpu.

```python
!python train.py --img 416 --batch 8 --epochs 50 --data data.yaml --cfg models/yolov5s.yaml --weights best.pt --name robo4_epoch150_s --adam
```

After training we need to test our model. We can do this in multiple ways. We can use images videos or web cam to check if the person is wearing a mak or not. The following are the : 
   1. weights: weights of the trained model
   2. source: input file/folder to run inference on, 0 for webcam
   3. output: directory to save results
   4. iou-thres: IOU threshold for NMS, defaults to 0.45
   5. conf-thres: object confidence threshold
